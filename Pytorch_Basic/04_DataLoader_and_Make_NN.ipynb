```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable
from torch.utils.data import Dataset, DataLoader

import numpy as np
```

```python
import torchvision
import torchvision.transforms as transforms
```

```python
transform = transforms.Compose([transforms.ToTensor(), #가져온 이미지를 Tensor 형테로 변환(np를 Tensor로) 채널이 변한다.
                                #ToTensor C x H x W 을 H x W x C로 바꿔준다.
                               transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])
#transforms은 선처리를 위한 함수. #CenterCrop 등 다른 함수도 있다.
#transforms.Compose는 선처리를 위한 컨테이너
```

```python
trainset = torchvision.datasets.CIFAR10(root="./data",
                                       train=True, #True로 하면 train 데이터로 가져온다.
                                       download=True,
                                       transform=transform)

testset = torchvision.datasets.CIFAR10(root="./data",
                                       train=False, #false로 하면 test 데이터로 가져온다.
                                       download=True,
                                       transform=transform)
#이미 선언되어 있는 데이터 셋을 가져온다. .을 붙여줘야 된다(안 붙이면 pemission denied).
```

```python
trainloader = DataLoader(trainset, batch_size=8, shuffle=True, num_workers=2)
testloader = DataLoader(testset, batch_size=8, shuffle=True, num_workers=2)
#num_workers는 CPU 사용 설정. batch_size가 클 수록 num_workers가 큰 게 좋다.
```

## Define the Network

- class 선언으로 사용하고자 하는 Network를 직접 구현할 수 있다.
- class로 선언할 Network의 필수 요소는
    - \__init__(self) : 사용할 함수들을 정의
    - forward(self, x) : 함수들을 사용하여 Network의 forward를 정의


```python
#먼저 이미지의 size를 확인해 보는 것이 좋다. 여기서는 [8, 3, 32, 32, 32] (batch가 8, channel이 3, width 32, height 32)

net = nn.Conv2d(3, 5, 5) #여기서 input은 channel의 수와 같다(채널). #channel 3 개, 필터 5개, 5x5 필터
out1 = net(imgs)
print(out1.shape) #5 x 5로 필터를 설정했으므로 output은 28 x 28이 된다.
```

```python
net2 = nn.Conv2d(5, 10, 5) #channel 5 개, 필터 10개, 5x5 필터
out2 = net2(out1)
print(out2.shape)
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.net1 = nn.Conv2d(3, 5, 5)
        self.net2 = nn.Conv2d(5, 10, 5)

    def forward(self, x):
        x = self.net1(x)
        x = self.net2(x)
        return x
```

```python
imgs = 0
for n, (img, labels) in enumerate(trainloader):
    print(n, img.shape, labels.shape)
    imgs = img
    break
```

```python
model = Model()

out = model(imgs)
print(out.shape)
```
